```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix , classification_report
%matplotlib inline
```

```python
df = pd.read_csv('customer_churn.csv')
df.head()
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.TotalCharges.dropna(inplace=True)
```

```python
def unique_values_of_col(df):
  for column in df:
    if df[column].dtypes=='object':

      print(f'{column}: {df[column].unique()}')
```

```python
unique_values_of_col(df)
```

```python
df['MultipleLines'].replace('No phone service', 'No', inplace=True)
df['OnlineSecurity'].replace('No internet service', 'No', inplace=True)
df['DeviceProtection'].replace('No internet service', 'No', inplace=True)
df['TechSupport'].replace('No internet service', 'No', inplace=True)
df['StreamingTV'].replace('No internet service', 'No', inplace=True)
df['StreamingMovies'].replace('No internet service', 'No', inplace=True)
df['OnlineBackup'].replace('No internet service', 'No', inplace=True)
```

```python
unique_values_of_col(df)
```

```python
encoding_1_0 = ['Churn', 'PaperlessBilling', 'StreamingMovies', 'StreamingTV', 'TechSupport', 'DeviceProtection',
                'OnlineBackup', 'OnlineSecurity', 'MultipleLines', 'PhoneService', 'Dependents', 'Partner']
```

```python
for col in encoding_1_0:
  df[col] = df[col].replace({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
df['gender'].replace({'Male': 1, 'Female': 0}, inplace=True)
```

```python
df2 = pd.get_dummies(data=df, columns = ['InternetService', 'Contract','PaymentMethod'], dtype='int')
df2.head()
```

```python
df2.dtypes
```

```python
scaler = MinMaxScaler()
df2[['tenure','MonthlyCharges', 'TotalCharges' ]] = scaler.fit_transform(df2[['tenure','MonthlyCharges', 'TotalCharges' ]])
df2.head()
```

```python
df2.describe()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
```

```python
X = df2.drop('Churn', axis=1)
y = df2['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model = keras.Sequential([
        keras.layers.Dense(26, input_dim=26, activation='relu'),
        keras.layers.Dropout(rate=0.3),
        keras.layers.Dense(25, activation='relu'),
        keras.layers.Dropout(rate=0.3),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dropout(rate=0.3),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])

    if weights == -1:
        model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight = weights)

    print(model.evaluate(X_test, y_test))

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)

    print("Classification Report: \n", classification_report(y_test, y_preds))

    return y_preds
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

**As we can see from the above estimations,our dataset has really low values of precision, recall and f1-score for 1, because we have imbalanced column values**

**SMOTE- this method is used to balnced the dataset by creting synthetically similar values with k-nearest neighbours**

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

y_sm.value_counts()
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
```

```python
y_train2.value_counts()
```

```python
y_preds = ANN(X_train2, y_train2, X_test2, y_test2, 'binary_crossentropy', -1)
```
